```python
class HealthPredictor:
    def __init__(self, symptoms_dict, diseases_list, description, symtoms, medications, precautions, diets, workout):
        self.symptoms_dict = symptoms_dict
        self.diseases_list = diseases_list
        self.description = description
        self.symtoms = symtoms
        self.medications = medications
        self.precautions = precautions
        self.diets = diets
        self.workout = workout

    def predict_disease(self, symptoms):
        num_features = 132
        feature_vector = [0] * num_features

        for symptom in symptoms:
            if symptom in self.symptoms_dict:
                feature_vector[self.symptoms_dict[symptom]] = 1

        x = svc_model.predict([feature_vector])
        predicted_disease = self.diseases_list[x[0]]

        return predicted_disease, x[0]

    def collect_symptoms(self):
        symptoms = []
        while True:
            symptom = input("Enter a symptom (or type 'done' to finish): ").strip().lower()
            if symptom == 'done':
                break
            symptoms.append(symptom)
        return symptoms

    def print_information(self, predicted_disease):
        predicted_description_info = self.description[self.description['Disease'] == predicted_disease]['Description'].values

        predicted_symptoms_info = self.symtoms[self.symtoms['Disease'] == predicted_disease][['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4']].values
        predicted_symptoms_info = pd.unique(predicted_symptoms_info.ravel())
        predicted_symptoms_info = pd.Series(predicted_symptoms_info).dropna().tolist()

        predicted_medications_info = self.medications[self.medications['Disease'] == predicted_disease]['Medication'].values

        predicted_precautions_info = self.precautions[self.precautions['Disease'] == predicted_disease][['Precaution_1', 'Precaution_2', 'Precaution_3','Precaution_4']].values
        predicted_precautions_info = pd.unique(predicted_precautions_info.ravel())
        predicted_precautions_info = pd.Series(predicted_precautions_info).dropna().tolist()

        predicted_diet_info = self.diets[self.diets['Disease'] == predicted_disease]['Diet'].values

        predicted_workout_info = self.workout[self.workout['disease'] == predicted_disease]['workout'].values

        print("Description of", predicted_disease, ":", predicted_description_info)
        print("")
        print("Symptoms:", predicted_symptoms_info)
        print("")
        print("Medications:", predicted_medications_info)
        print("")
        print("Precautions:", predicted_precautions_info)
        print("")
        print("Diet:", predicted_diet_info)
        print("")
        print("Workout:", predicted_workout_info)

# Instancier la classe HealthPredictor avec les dictionnaires de donnÃ©es
health_predictor = HealthPredictor(symptoms_dict, diseases_list, description, symtoms, medications, precautions, diets, workout)

# Collecter les symptÃ´mes de l'utilisateur
user_symptoms = health_predictor.collect_symptoms()

# PrÃ©dire la maladie basÃ©e sur les symptÃ´mes
predicted_disease, disease_code = health_predictor.predict_disease(user_symptoms)

# Imprimer les informations sur la maladie prÃ©dite
health_predictor.print_information(predicted_disease)
```
